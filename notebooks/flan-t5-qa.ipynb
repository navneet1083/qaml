```python
from datasets import load_dataset
from transformers import AutoModelWithLMHead, pipeline, AutoModelForSeq2SeqLM
from transformers import AutoTokenizer
from transformers import GenerationConfig
```

```python
t5_qa_model = AutoModelForSeq2SeqLM.from_pretrained("google/t5-small-ssm-nq")
t5_tok = AutoTokenizer.from_pretrained("google/t5-small-ssm-nq")
```

```python
?t5_tok
```

```python
# input_ids = t5_tok("When was Franklin D. Roosevelt born?", return_tensors="pt").input_ids
input_ids = t5_tok("What is Pan card ?", return_tensors="pt").input_ids
# gen_output = t5_qa_model.generate(input_ids)[0]
gen_output = t5_qa_model.generate(input_ids)

preds = [
    t5_tok.decode(gen_id, skip_special_tokens=True, max_new_tokens=20)
    for gen_id in gen_output
]

# print(t5_tok.decode(gen_output, skip_special_tokens=True, max_new_tokens=20))
"".join(preds)
```
